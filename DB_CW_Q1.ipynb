```r
# Example data
x <- 1:10
y <- rnorm(10)
# Perform correlation test
cor_test <- cor.test(x, y)
# Extracting correlation test results
cor_results <- data.frame( Estimate = cor_test$estimate, p_value = cor_test$p.value, Confidence_Interval = paste(cor_test$conf.int[1], cor_test$conf.int[2], sep = " to ") )
# Displaying results as a table print(cor_results)
```

```r
# Load necessary libraries
library(ggplot2)

# 1. Scatter Plot
# Scatter plot between "Relative Compactness" and "Heating Load"
scatter_plot <- ggplot(data, aes(x = X1, y = Y1)) +
  geom_point() +
  labs(title = "Scatter Plot: Relative Compactness vs. Heating Load")

# Display the plots
scatter_plot
```

```r
scatter_plot2 <- ggplot(data, aes(x = X1, y = Y2)) +
  geom_point() +
  labs(title = "Scatter Plot: Relative Compactness vs. Cooling Load")
scatter_plot2
```

```r
# 2. Histogram
# Histogram of "Heating Load"
histogram_heating <- ggplot(data, aes(x = Y1)) +
  geom_histogram(bins = 10, fill = "skyblue", color = "black") +
  labs(title = "Histogram: Heating Load", x = "Heating Load", y = "Frequency")

histogram_heating
```

```r
# Histogram of "Cooling Load"
histogram_cooling <- ggplot(data, aes(x = Y2)) +
  geom_histogram(bins = 10, fill = "lightgreen", color = "black") +
  labs(title = "Histogram: Cooling Load", x = "Cooling Load", y = "Frequency")

histogram_cooling
```

```r
# 3. Bar Plot
# Average heating load across different orientations
avg_heating_orientation <- aggregate(Y1 ~ X6, data = data, FUN = mean)
barplot_heating_orientation <- ggplot(avg_heating_orientation, aes(x = factor(X6), y = Y1)) +
  geom_bar(stat = "identity", fill = "orange") +
  labs(title = "Bar Plot: Average Heating Load across Orientations", x = "Orientation", y = "Average Heating Load")

barplot_heating_orientation
```

```r
# Average cooling load across different orientations
avg_cooling_orientation <- aggregate(Y2 ~ X6, data = data, FUN = mean)
barplot_cooling_orientation <- ggplot(avg_cooling_orientation, aes(x = factor(X6), y = Y2)) +
  geom_bar(stat = "identity", fill = "lightblue") +
  labs(title = "Bar Plot: Average Cooling Load across Orientations", x = "Orientation", y = "Average Cooling Load")


barplot_cooling_orientation
```
