```r
new_record <- data.frame(
  X1 = 0.80,
  X2 = 500.5,
  X3 = 300.0,
  X4 = 120.25,
  X5 = 10.0,
  X6 = 3,
  X7 = 0.0,
  X8 = 0,
  Y1 = 20.55,
  Y2 = 28.33
)

# Append the new record to the existing dataframe
data <- rbind(data, new_record)
```

```r
result <- sqldf("SELECT * FROM data where X2 = 500.5")
result
```

```r
library(ggplot2)
library(plotly)
library(dplyr)
```

```r
filtered_data <- data %>%
  filter(X5 == 7) %>%
  mutate(Surface_Area_Percentage = X2 / sum(X2) * 100)
```

```r
filtered_data
```

```r
# Example data
x <- 1:10
y <- rnorm(10)
# Perform correlation test
cor_test <- cor.test(x, y)
# Extracting correlation test results
cor_results <- data.frame( Estimate = cor_test$estimate, p_value = cor_test$p.value, Confidence_Interval = paste(cor_test$conf.int[1], cor_test$conf.int[2], sep = " to ") )
# Displaying results as a table print(cor_results)
```

```r
ggplot(data, aes(x = X1, y = Y1)) +
  geom_point() +
  labs(title = "X1 vs Y1", x = "X1", y = "Y1")
```

```r
# Line plot
ggplot(data, aes(x = X2, y = Y1)) +
  geom_line() +
  labs(title = "Line Plot", x = "X1", y = "Y1")
```

```r
# Bar plot
ggplot(data, aes(x = X3, fill = Y2)) +
  geom_bar() +
  labs(title = "Bar Plot", x = "X3", y = "Count")
```

```r
# Histogram
ggplot(data, aes(x = X4)) +
  geom_histogram() +
  labs(title = "Histogram", x = "X1", y = "Frequency")
```

```r
# Box plot
ggplot(data, aes(x = Y1, y = X1)) +
  geom_boxplot() +
  labs(title = "Box Plot", x = "Y1", y = "X1")
```
